# Model fit during experiments

## Hyperparamters

```python
params_stress = []
params_resist = []
params_obj = []
for k in range(total_exp):
    with open(RESULT_DIR + "/" + str(k) + '_gp_objects.pkl', 'rb') as outp:
        data = pickle.load(outp)
        GP_stress = data['stress']
        params_stress.append(GP_stress.kernel.theta)
        
        GP_resist = data['resist']
        params_resist.append(GP_resist.kernel.theta)
        
        GP_obj = data['obj']
        params_obj.append(GP_obj.kernel.theta)

params_stress = np.array(params_stress)
params_resist = np.array(params_resist)
params_obj = np.array(params_obj)

fig, axs = plt.subplots(1,3, figsize=(20, 3))
axs[0].plot(range(1, total_exp+1), params_stress[:, 0], "-*", label="stress")
axs[0].plot(range(1, total_exp+1), params_resist[:, 0], "-*", label="resist")
axs[0].plot(range(1, total_exp+1), params_obj[:, 0], "-*", label="obj")
axs[0].set_title("Sigma used during predictions")

axs[1].plot(range(1, total_exp+1), params_stress[:, 1], "-*", label="stress")
axs[1].plot(range(1, total_exp+1), params_resist[:, 1], "-*", label="resist")
axs[1].plot(range(1, total_exp+1), params_obj[:, 1], "-*", label="obj")
axs[1].set_title("Length scale in pressure directions used during fit")
axs[1].set_yscale('log')

axs[2].plot(range(1, total_exp+1), params_stress[:, 2], "-*", label="stress")
axs[2].plot(range(1, total_exp+1), params_resist[:, 2], "-*", label="resist")
axs[2].plot(range(1, total_exp+1), params_obj[:, 2], "-*", label="obj")
axs[2].set_title("Length scale in power directions used during fit")
axs[2].set_yscale('log')

[ax.set_xlabel('runs') for ax in axs]
[ax.legend() for ax in axs]
plt.show()
```

## Log likelihood

```python
nll_stress = []
nll_resist = []
nll_obj = []
for k in range(total_exp):
    with open(RESULT_DIR + "/" + str(k) + '_gp_objects.pkl', 'rb') as outp:
        data = pickle.load(outp)
        GP_stress = data['stress']
        nll_stress.append(-GP_stress.negative_log_likelihood(GP_stress.kernel.theta))
        
        GP_resist = data['resist']
        nll_resist.append(-GP_resist.negative_log_likelihood(GP_stress.kernel.theta))
        
        GP_obj = data['obj']
        nll_obj.append(-GP_obj.negative_log_likelihood(GP_stress.kernel.theta))
        
fig, axs = plt.subplots(1,3, figsize=(20, 3))
axs[0].plot(range(1, total_exp+1), nll_stress, "-*", label="stress")
axs[0].set_title("Stress model")
axs[1].plot(range(1, total_exp+1), nll_resist, "-*", label="resist")
axs[1].set_title("Resistance model")
axs[2].plot(range(1, total_exp+1), nll_obj, "-*", label="obj")
axs[2].set_title("Objective model")

[ax.set_xlabel('runs') for ax in axs]
[ax.set_ylabel('log likelihood') for ax in axs]
[ax.set_yscale('symlog') for ax in axs]
plt.show()
```

## Condition number of correlation matrix

```python
stress_condition_number = []
resist_condition_number = []
obj_condition_number = []
stress_norm = []
resist_norm = []
obj_norm = []
for k in range(total_exp):
    with open(RESULT_DIR + "/" + str(k) + '_gp_objects.pkl', 'rb') as outp:
        data = pickle.load(outp)
        GP_stress = data['stress']
        stress_condition_number.append(np.linalg.cond(GP_stress.K11))
        stress_norm.append(np.linalg.norm(GP_stress.K11))
        
        GP_resist = data['resist']
        resist_condition_number.append(np.linalg.cond(GP_resist.K11))
        resist_norm.append(np.linalg.norm(GP_resist.K11))
        
        GP_obj = data['obj']
        obj_condition_number.append(np.linalg.cond(GP_obj.K11))
        obj_norm.append(np.linalg.norm(GP_obj.K11))
        
fig, axs = plt.subplots(1,3, figsize=(20, 3))
axs[0].plot(range(1, total_exp+1), stress_condition_number, "-*", label="stress")
axs[0].set_title("Stress model")
axs[1].plot(range(1, total_exp+1), resist_condition_number, "-*", label="resist")
axs[1].set_title("Resistance model")
axs[2].plot(range(1, total_exp+1), obj_condition_number, "-*", label="obj")
axs[2].set_title("Objective model")

[ax.set_xlabel('runs') for ax in axs]
[ax.set_ylabel('condition number') for ax in axs]
[ax.set_yscale('log') for ax in axs]
plt.show()

fig, axs = plt.subplots(1,3, figsize=(20, 3))
axs[0].plot(range(1, total_exp+1), stress_norm, "-*", label="stress")
axs[0].set_title("Stress model")
axs[1].plot(range(1, total_exp+1), resist_norm, "-*", label="resist")
axs[1].set_title("Resistance model")
axs[2].plot(range(1, total_exp+1), obj_norm, "-*", label="obj")
axs[2].set_title("Objective model")

[ax.set_xlabel('runs') for ax in axs]
[ax.set_ylabel('norm') for ax in axs]
[ax.set_yscale('log') for ax in axs]
plt.show()
```

A matrix has very high condition number means that the matrix is nearly singular. 
This, in turn, implies that one or more columns are close to linear combinations of the rest of the columns.

# Objective function and Acquisition function analysis

The max value of the fit, sd of the fit and actual objective function value at the suggested point

```python
obj_gp_mu_before = []
obj_gp_mu_after = []
obj_gp_sd_before = []
obj_gp_sd_after = []
obj_value_before = []
obj_value_after = []

for k in range(total_exp-1):
    x_suggested = np.load(RESULT_DIR+"/"+str(k)+'_x_next.npy')

    # How old model performed at the experiment location
    with open(RESULT_DIR + "/" + str(k) + '_gp_objects.pkl', 'rb') as file:
        data = pickle.load(file)
        GP_obj = data['obj']   
        mu, sd = GP_obj.posterior(x_suggested)

        obj_gp_mu_before.append(mu.reshape(-1)) 
        obj_gp_sd_before.append(sd.reshape(-1))
        
        # ------------------------
        GP_stress = data['stress']
        stress_mu, _, stress_mu_grad = GP_stress.posterior(x_suggested, calc_grad=True)
        
        GP_resist = data['resist']
        resist_mu, _ = GP_resist.posterior(x_suggested)        
        obj_val, _, _, _, _  = stress_obj(stress_mu, stress_mu_grad, resist_mu)
        obj_value_before.append(obj_val[0][0])

    # how new model performed at the suggested location after new data point
    with open(RESULT_DIR + "/" + str(k+1) + '_gp_objects.pkl', 'rb') as file:
        data = pickle.load(file)
        GP_obj = data['obj']   
        mu, sd = GP_obj.posterior(x_suggested)

        obj_gp_mu_after.append(mu.reshape(-1)) 
        obj_gp_sd_after.append(sd.reshape(-1)) 

        # ------------------------
        GP_stress = data['stress']
        stress_mu, _, stress_mu_grad = GP_stress.posterior(x_suggested, calc_grad=True)
        
        GP_resist = data['resist']
        resist_mu, _ = GP_resist.posterior(x_suggested)        
        obj_val, _, _, _, _  = stress_obj(stress_mu, stress_mu_grad, resist_mu)
        obj_value_after.append(obj_val[0][0])   
        
obj_gp_mu_before = np.array(obj_gp_mu_before)
obj_gp_sd_before = np.array(obj_gp_sd_before)
obj_gp_mu_after = np.array(obj_gp_mu_after)
obj_gp_sd_after = np.array(obj_gp_sd_after)
obj_value_before = np.array(obj_value_before)
obj_value_after = np.array(obj_value_after)

fig, ax = plt.subplots(1,3,figsize=[15,4])
ax = ax.reshape(-1) 
exp_indx = range(1,total_exp)

ax[0].plot(exp_indx, obj_gp_mu_before, "-*", label='before', lw=4)
ax[0].plot(exp_indx, obj_gp_mu_after, "-*", label='after')
ax[0].set_title("GP mean at suggested location")

ax[1].plot(exp_indx, obj_gp_sd_before, "-*", label='before', lw=4)
ax[1].plot(exp_indx, obj_gp_sd_after, "-*", label='after')
ax[1].set_title("GP SD at suggested location")

ax[2].plot(exp_indx, obj_value_before, "-*", label='before', lw=4)
ax[2].plot(exp_indx, obj_value_after, "-*", label='after')
ax[2].set_title("Objective value")

[axs.set_xlabel("runs") for axs in ax]
[axs.set_xticks(exp_indx) for axs in ax]
[axs.set_ylabel("Values") for axs in ax]
[axs.legend() for axs in ax]
fig.tight_layout()
plt.show()
```

The above plots shows how the GP models were updated after updating observations at the suggested points, not much conclusions to make, except that it is working well

```python
obj_gp_mu_before = []
obj_gp_mu_after = []
obj_gp_sd_before = []
obj_gp_sd_after = []
obj_value_before = []
obj_value_after = []

for k in range(total_exp-1):
    x_suggested = np.load(RESULT_DIR+"/"+str(k)+'_x_next.npy')

    # How old model performed at the experiment location
    with open(RESULT_DIR + "/" + str(k) + '_gp_objects.pkl', 'rb') as file:
        data = pickle.load(file)
        GP_obj = data['obj']   
        mu, sd = GP_obj.posterior(x_suggested)

        obj_gp_mu_before.append(mu.reshape(-1)) 
        obj_gp_sd_before.append(sd.reshape(-1))
        
        # ------------------------
        GP_stress = data['stress']
        stress_mu, _, stress_mu_grad = GP_stress.posterior(x_suggested, calc_grad=True)
        
        GP_resist = data['resist']
        resist_mu, _ = GP_resist.posterior(x_suggested)        
        obj_val, _, _, _, _  = stress_obj(stress_mu, stress_mu_grad, resist_mu)
        obj_value_before.append(obj_val[0][0])

    # how new model performed at the suggested location after new data point
    with open(RESULT_DIR + "/" + str(total_exp-1) + '_gp_objects.pkl', 'rb') as file:
        data = pickle.load(file)
        GP_obj = data['obj']   
        mu, sd = GP_obj.posterior(x_suggested)

        obj_gp_mu_after.append(mu.reshape(-1)) 
        obj_gp_sd_after.append(sd.reshape(-1)) 

        # ------------------------
        GP_stress = data['stress']
        stress_mu, _, stress_mu_grad = GP_stress.posterior(x_suggested, calc_grad=True)
        
        GP_resist = data['resist']
        resist_mu, _ = GP_resist.posterior(x_suggested)        
        obj_val, _, _, _, _  = stress_obj(stress_mu, stress_mu_grad, resist_mu)
        obj_value_after.append(obj_val[0][0])   
        
obj_gp_mu_before = np.array(obj_gp_mu_before)
obj_gp_sd_before = np.array(obj_gp_sd_before)
obj_gp_mu_after = np.array(obj_gp_mu_after)
obj_gp_sd_after = np.array(obj_gp_sd_after)
obj_value_before = np.array(obj_value_before)
obj_value_after = np.array(obj_value_after)

fig, ax = plt.subplots(1,3,figsize=[15,4])
ax = ax.reshape(-1) 
exp_indx = range(1,total_exp)

ax[0].plot(exp_indx, obj_gp_mu_before, "-*", label='before', lw=4)
ax[0].plot(exp_indx, obj_gp_mu_after, "-*", label='final')
ax[0].set_title("GP mean at suggested location")

ax[1].plot(exp_indx, obj_gp_sd_before, "-*", label='before', lw=4)
ax[1].plot(exp_indx, obj_gp_sd_after, "-*", label='final')
ax[1].set_title("GP SD at suggested location")

ax[2].plot(exp_indx, obj_value_before, "-*", label='before', lw=4)
ax[2].plot(exp_indx, obj_value_after, "-*", label='final')
ax[2].set_title("Objective value")

[axs.set_xlabel("runs") for axs in ax]
[axs.set_xticks(exp_indx) for axs in ax]
[axs.set_ylabel("Values") for axs in ax]
[axs.legend() for axs in ax]
fig.tight_layout()
plt.show()
```
